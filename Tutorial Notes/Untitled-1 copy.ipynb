
# **Power System Feasibility Analysis**

---

## **Problem Statement**
We are given a **three-bus power system** where power transactions take place between different nodes (buses). Our goal is to verify whether the given sets of power transactions are **simultaneously feasible** by checking whether the power flows in the transmission lines remain **within their thermal capacity limits**.

---

## **Step 1: Understanding the Power System**

The power system consists of:
- **Three buses (nodes):** Representing locations where power is injected or drawn.
- **Three transmission lines:** Connecting the buses with specific **reactance values**.

### **System Constraints**
1. The **thermal capacity limit** of each transmission line is:

   $$
   P^{\max} = 250 \text{ MW}
   $$

2. The **line reactances** (impedance values) are:

   $$
   x_{12} = 0.2, \quad x_{13} = 0.4, \quad x_{23} = 0.4
   $$

3. **Power balance equations** are derived using **Kirchhoff's Current Law (KCL)**.
4. **Loop equation** is derived using **Kirchhoff's Voltage Law (KVL)**.

---

## **Step 2: Deriving the Matrix Equation**
We derive the power balance equations using **Kirchhoff‚Äôs Current Law (KCL)** and **Kirchhoff‚Äôs Voltage Law (KVL)**.

### **Power Balance at Each Bus (KCL)**
**Kirchhoff‚Äôs Current Law (KCL)** states that the total power injected into a node must be equal to the total power leaving the node.

- **Bus 1:**
  
  The net power injected at Bus 1 must be equal to the sum of the power flows on lines **1-2** and **1-3**:
  
  $$
  P_{\text{injected,1}} - P_{\text{demand,1}} = F_{12} + F_{13}
  $$

- **Bus 2:**
  
  The net power injected at Bus 2 must be equal to the sum of power flowing out through line **2-3**, while power is received from line **1-2**:
  
  $$
  P_{\text{injected,2}} - P_{\text{demand,2}} = -F_{12} + F_{23}
  $$

- **Bus 3:**
  
  The net power injected at Bus 3 must be equal to the sum of power flows on **1-3** and **2-3**:
  
  $$
  P_{\text{injected,3}} - P_{\text{demand,3}} = -F_{13} - F_{23}
  $$

### **Loop Equation (KVL)**
**Kirchhoff‚Äôs Voltage Law (KVL)** states that the sum of voltage drops around any closed loop must be zero.

Using **Ohm‚Äôs Law**, the voltage drop across each transmission line is proportional to the power flow divided by the line reactance:

$$
\frac{F_{12}}{x_{12}} + \frac{F_{23}}{x_{23}} - \frac{F_{13}}{x_{13}} = 0
$$

Given the reactance values:

$$
x_{12} = 0.2, \quad x_{13} = 0.4, \quad x_{23} = 0.4
$$

Substituting:

$$
\frac{F_{12}}{0.2} + \frac{F_{23}}{0.4} - \frac{F_{13}}{0.4} = 0
$$

Multiplying by **0.2** to clear fractions:

$$
0.2 F_{12} - 0.4 F_{13} + 0.4 F_{23} = 0
$$

Thus, the third row in the matrix equation corresponds to:

$$
\begin{bmatrix} 0.2 & -0.4 & 0.4 \end{bmatrix}
$$

---

## **Step 3: Matrix Representation**
Combining all equations, we write the system in matrix form:

$$
\begin{bmatrix}
1 & 1 & 0 \\
-1 & 0 & 1 \\
0.2 & -0.4 & 0.4
\end{bmatrix}
\begin{bmatrix}
F_{12} \\
F_{13} \\
F_{23}
\end{bmatrix}
=
\begin{bmatrix}
P_{\text{injected,1}} - P_{\text{demand,1}} \\
P_{\text{injected,2}} - P_{\text{demand,2}} \\
0
\end{bmatrix}
$$
