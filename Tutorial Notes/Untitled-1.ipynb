# **Power System Feasibility Analysis**

## **Problem Statement**
We are given a **three-bus power system** where power transactions take place between different nodes (buses). Our goal is to verify whether the given sets of power transactions are **simultaneously feasible** by checking whether the power flows in the transmission lines remain **within their thermal capacity limits**.

---

## **Step 1: Understanding the Power System**

The power system consists of:
- **Three buses (nodes):** Representing locations where power is injected or drawn.
- **Three transmission lines:** Connecting the buses with specific **reactance values**.

### **System Constraints**
1. The **thermal capacity limit** of each transmission line is:

   $$ P^{\max} = 250 \text{ MW} $$

2. The **line reactances** (impedance values) are:

   $$
   x_{12} = 0.2, \quad x_{13} = 0.4, \quad x_{23} = 0.4
   $$

3. **Power balance equations** are derived using Kirchhoff's Current Law (KCL).
4. **Loop equation** is derived using Kirchhoff's Voltage Law (KVL).

---

## **Step 2: Writing the Power Balance Equations**

Using **KCL**, the power balance at each bus is:

- **Bus 1 (B):**
  
  The net power injected at bus 1 must be equal to the sum of the power flows on lines **1-2** and **1-3**:
  
  $$ B - Y = F_{12} + F_{13} $$

- **Bus 2 (A):**
  
  The net power injected at bus 2 must be equal to the sum of power flowing out through line **2-3**, while power is received from line **1-2**:
  
  $$ A - X = -F_{12} + F_{23} $$

- **Bus 3 (C):**
  
  The net power injected at bus 3 must be equal to the sum of power flows on **1-3** and **2-3**:
  
  $$ C - Z = -F_{13} - F_{23} $$

### **Loop Equation (KVL)**
Applying **KVL** around the loop:

$$ F_{12} x_{12} + F_{23} x_{23} - F_{13} x_{13} = 0 $$

This equation ensures that the power flows satisfy the system constraints.

---

## **Step 3: Solving for Different Transaction Sets**
We solve the system of equations for each transaction set and verify whether power flows remain **within the 250 MW limit**.

---

# **Power Flow Equations Using Kirchhoff’s Laws**

---

## **Step 1: Understanding the Power System**
The power system consists of:
- **Three buses (nodes)**: Where power is either injected (generation) or withdrawn (load).
- **Three transmission lines**: Connecting the buses.

Each line carries a power flow:
- $F_{12}$ (Power flow from Bus 1 to Bus 2)
- $F_{13}$ (Power flow from Bus 1 to Bus 3)
- $F_{23}$ (Power flow from Bus 2 to Bus 3)

The **transaction set** represents the power exchange:

| **Seller** | **Buyer** | **Amount (MW)** |
|------------|----------|----------------|
| B          | X        | 200            |
| A          | Z        | 400            |
| C          | Y        | 300            |

Each bus follows the **power balance equation**, derived using **Kirchhoff’s Current Law (KCL)**.

---

## **Step 2: Writing Kirchhoff’s Current Law (KCL) Equations**
Applying **KCL**, the sum of power flows at each bus must be zero.

---

### **Equation at Bus 1**
Power injected at **Bus 1**:

$$
P_{\text{injected,1}} - P_{\text{demand,1}} = F_{12} + F_{13}
$$

$$
200 - 300 = F_{12} + F_{13}
$$

$$
-100 = F_{12} + F_{13}
$$

Thus, the first row in the matrix equation corresponds to:

$$
\begin{bmatrix} 1 & 1 & 0 \end{bmatrix}
$$

---

### **Equation at Bus 2**
Power injected at **Bus 2**:

$$
P_{\text{injected,2}} - P_{\text{demand,2}} = -F_{12} + F_{23}
$$

$$
400 - 200 = -F_{12} + F_{23}
$$

$$
200 = -F_{12} + F_{23}
$$

Thus, the second row in the matrix equation corresponds to:

$$
\begin{bmatrix} -1 & 0 & 1 \end{bmatrix}
$$

---

## **Step 3: Writing Kirchhoff’s Voltage Law (KVL) Equation**
Applying **KVL**, the sum of voltage drops around a closed loop must be zero:

$$
\frac{F_{12}}{x_{12}} + \frac{F_{23}}{x_{23}} - \frac{F_{13}}{x_{13}} = 0
$$

Given the reactance values:

$$
x_{12} = 0.2, \quad x_{13} = 0.4, \quad x_{23} = 0.4
$$

Substituting:

$$
\frac{F_{12}}{0.2} + \frac{F_{23}}{0.4} - \frac{F_{13}}{0.4} = 0
$$

Multiplying by $0.2$ to simplify:

$$
0.2 F_{12} - 0.4 F_{13} + 0.4 F_{23} = 0
$$

Thus, the third row in the matrix equation corresponds to:

$$
\begin{bmatrix} 0.2 & -0.4 & 0.4 \end{bmatrix}
$$

---



## **Expanded Solution**

We are given the following system of equations:

$$
\begin{bmatrix}
1 & 1 & 0 \\
-1 & 0 & 1 \\
0.2 & -0.4 & 0.4
\end{bmatrix}
\begin{bmatrix}
F_{12} \\
F_{13} \\
F_{23}
\end{bmatrix}
=
\begin{bmatrix}
200 - 300 \\
400 - 200 \\
0
\end{bmatrix}
$$

### **Step 1: Simplify the Right-Hand Side**

$$
\begin{bmatrix}
200 - 300 \\
400 - 200 \\
0
\end{bmatrix}
=
\begin{bmatrix}
-100 \\
200 \\
0
\end{bmatrix}
$$

Thus, the system becomes:

$$
\begin{bmatrix}
1 & 1 & 0 \\
-1 & 0 & 1 \\
0.2 & -0.4 & 0.4
\end{bmatrix}
\begin{bmatrix}
F_{12} \\
F_{13} \\
F_{23}
\end{bmatrix}
=
\begin{bmatrix}
-100 \\
200 \\
0
\end{bmatrix}
$$

### **Step 2: Convert to Individual Equations**
Expanding the matrix multiplication:

1. $ 1F_{12} + 1F_{13} + 0F_{23} = -100 $

   $$
   F_{12} + F_{13} = -100
   $$

2. $ -1F_{12} + 0F_{13} + 1F_{23} = 200 $

   $$
   -F_{12} + F_{23} = 200
   $$

3. $ 0.2F_{12} - 0.4F_{13} + 0.4F_{23} = 0 $

   $$
   0.2 F_{12} - 0.4 F_{13} + 0.4 F_{23} = 0
   $$

### **Step 3: Solve for $F_{12}$, $F_{13}$, and $F_{23}$**

#### Solve for $F_{12}$ and $F_{13}$
From equation (1):

$$
F_{12} = -100 - F_{13}
$$

Substituting into equation (2):

$$
-(-100 - F_{13}) + F_{23} = 200
$$

$$
100 + F_{13} + F_{23} = 200
$$

$$
F_{13} + F_{23} = 100
$$

From equation (3):

$$
0.2(-100 - F_{13}) - 0.4 F_{13} + 0.4 F_{23} = 0
$$

$$
-20 - 0.2 F_{13} - 0.4 F_{13} + 0.4 F_{23} = 0
$$

$$
-20 - 0.6 F_{13} + 0.4 F_{23} = 0
$$

$$
0.4 F_{23} = 20 + 0.6 F_{13}
$$

$$
F_{23} = \frac{20 + 0.6 F_{13}}{0.4}
$$

$$
F_{23} = 50 + 1.5 F_{13}
$$

#### Solve for $F_{13}$
Substituting $F_{23} = 50 + 1.5 F_{13}$ into $F_{13} + F_{23} = 100$:

$$
F_{13} + (50 + 1.5 F_{13}) = 100
$$

$$
F_{13} + 50 + 1.5 F_{13} = 100
$$

$$
2.5 F_{13} = 50
$$

$$
F_{13} = 20
$$

#### Solve for $F_{12}$ and $F_{23}$
Using $F_{12} = -100 - F_{13}$:

$$
F_{12} = -100 - 20
$$

$$
F_{12} = -120
$$

Using $F_{23} = 50 + 1.5 F_{13}$:

$$
F_{23} = 50 + 1.5 (20)
$$

$$
F_{23} = 80
$$

### **Final Solution**
$$
F_{12} = -120 \text{ MW}, \quad |F_{12}| \leq 250 \text{ MW}
$$

$$
F_{13} = 20 \text{ MW}, \quad |F_{13}| \leq 250 \text{ MW}
$$

$$
F_{23} = 80 \text{ MW}, \quad |F_{23}| \leq 250 \text{ MW}
$$

Thus, the solution satisfies the given power flow constraints.



#### **Solution**
$$ F_{12} = -120 \text{ MW} \Rightarrow |F_{12}| \leq 250 \text{ MW} $$

$$ F_{13} = 20 \text{ MW} \Rightarrow |F_{13}| \leq 250 \text{ MW} $$

$$ F_{23} = 80 \text{ MW} \Rightarrow |F_{23}| \leq 250 \text{ MW} $$

✅ **These transactions are feasible.**
