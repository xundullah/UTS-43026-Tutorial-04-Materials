# Problem 5.3: Generator Revenues and Consumer Payments

## Question

Calculate the **generator revenues** and the **consumer payments** for all the cases considered in Problem 5.2.  
**Who benefits** from the line connecting these two buses?

---

## Definitions

Let:

- $\pi_A$, $\pi_B$: Electricity prices at buses A and B
- $D_A$, $D_B$: Demand at buses A and B
- $P_A$, $P_B$: Generation at buses A and B

---

## Revenue and Payment Formulas

### Consumer Payments:

- At bus A:  
  $$ E_A = D_A \cdot \pi_A $$

- At bus B:  
  $$ E_B = D_B \cdot \pi_B $$

### Generator Revenues:

- Generator A:  
  $$ R_A = P_A \cdot \pi_A $$

- Generator B:  
  $$ R_B = P_B \cdot \pi_B $$

---

### Given:

From **Problem 5.2**, we have:

- $D_A = 2000 \ \text{MW}$  
- $D_B = 1000 \ \text{MW}$

---

## (a) Line is disconnected

- $\pi_A = 80$, $\pi_B = 35$  
- $P_A = 2000$, $P_B = 1000$

**Consumer Payments:**

- $$ E_A = 2000 \cdot 80 = 160{,}000 $$
- $$ E_B = 1000 \cdot 35 = 35{,}000 $$

**Generator Revenues:**

- $$ R_A = 2000 \cdot 80 = 160{,}000 $$
- $$ R_B = 1000 \cdot 35 = 35{,}000 $$

---

## (b) Line in service with unlimited capacity

- $\pi_A = \pi_B = 53$  
- $P_A = 1100$, $P_B = 1900$

- $$ E_A = 2000 \cdot 53 = 106{,}000 $$
- $$ E_B = 1000 \cdot 53 = 53{,}000 $$

**Generator Revenues:**

- $$ R_A = 1100 \cdot 53 = 58{,}300 $$
- $$ R_B = 1900 \cdot 53 = 100{,}700 $$
