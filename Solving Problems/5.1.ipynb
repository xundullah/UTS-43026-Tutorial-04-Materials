We are given a 3-bus power system with thermal capacity constraints on each transmission line. The goal is to check whether given sets of transactions are **simultaneously feasible** considering these constraints.

### Transmission Lines and Limits:

- Bus 2â€“3: $x_{23} = 0.4 \, \text{p.u.}$, $P^{\text{MAX}} = 250 \, \text{MW}$

We denote:
- $F_{12}$: Power flow from bus 1 to 2  
- $F_{13}$: Power flow from bus 1 to 3  
- $F_{23}$: Power flow from bus 2 to 3

## ðŸ§® Step-by-Step Equations

### 1. **Power Balance Equations (KCL)**

- **Bus 1**: $P_B - P_Y = F_{12} + F_{13}$
- **Bus 2**: $P_A - P_X = -F_{12} + F_{23}$
- **Bus 3**: $P_C - P_Z = -F_{13} - F_{23}$

### 2. **Loop Equation (KVL)**

Using KVL around the loop:

$$
F_{12} \cdot x_{12} + F_{23} \cdot x_{23} - F_{13} \cdot x_{13} = 0
$$

Substituting values:

$$
F_{12} \cdot 0.2 + F_{23} \cdot 0.4 - F_{13} \cdot 0.4 = 0
$$
