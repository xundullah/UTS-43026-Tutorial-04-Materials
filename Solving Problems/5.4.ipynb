# Problem 5.4: Congestion Surplus for Case (e) of Problem 5.2

## üßÆ Definition

The **congestion surplus** is defined as the difference between:

- The **total consumer payments**, and  
- The **total generator revenues**

Mathematically:

$$
\text{cs} = (D_A \cdot \pi_A + D_B \cdot \pi_B) - (P_A \cdot \pi_A + P_B \cdot \pi_B)
$$

This can be rearranged by grouping:

$$
\text{cs} = (D_A - P_A) \cdot \pi_A + (D_B - P_B) \cdot \pi_B
$$

From the power balance equations:

- $D_A - P_A = -F_{AB}$
- $D_B - P_B = F_{AB}$

So we substitute:

$$
\text{cs} = F_{AB} \cdot (\pi_B - \pi_A)
$$

---

## ‚úÖ Apply to Case (e) of Problem 5.2

From case (e):

- $F_{AB} = P_A - D_A = 1400 - 2000 = -600 \ \text{MW}$
- $\pi_A = 62 \ \$/\text{MWh}$
- $\pi_B = 47 \ \$/\text{MWh}$

Then:

$$
\text{cs} = (-600) \cdot (47 - 62) = (-600) \cdot (-15) = 9000 \ \$
$$

---

## ‚úÖ Confirm with Problem 5.3

From Problem 5.3, case (e):

- $E_A = 124{,}000$
- $E_B = 47{,}000$
- $R_A = 86{,}800$
- $R_B = 75{,}200$

So:

$$
\text{cs} = (E_A + E_B) - (R_A + R_B) \\
\text{cs} = (124{,}000 + 47{,}000) - (86{,}800 + 75{,}200) = 171{,}000 - 162{,}000 = 9{,}000 \ \$
$$

‚úÖ The results match.

---

## üîç When is the congestion surplus zero?

Recall:

$$
\text{cs} = F_{AB} \cdot (\pi_B - \pi_A)
$$

So congestion surplus is **zero** when either:

1. $F_{AB} = 0$ (i.e. no power flows across the line)  
2. $\pi_A = \pi_B$ (i.e. no price difference between the buses)

‚úÖ This happens when:
- There is **no congestion**, and
- Power can freely flow to balance prices.
